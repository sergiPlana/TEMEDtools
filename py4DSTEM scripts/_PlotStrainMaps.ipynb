```python
#Load the data
from tifffile import imread
StoringName = '4DSTEM_QP_area5_84x87_10nmRes_CamLen400nm_PED0p5_binned_shifted_Yoffset_0_Ysize_87'
EXX = imread(StoringName+'_strain_raw_exx.tiff')
EYY = imread(StoringName+'_strain_raw_eyy.tiff')
EXY = imread(StoringName+'_strain_raw_exy.tiff')
Rot = imread(StoringName+'_strain_raw_rot.tiff')
```

```python
#Plotting parameters
import numpy as np
import matplotlib.pyplot as plt
titlesize = 20
colorbarLabelSize = 14
vrange_exx = [-1, 0]
vrange_eyy=[-0.5, 0.5]
vrange_exy = [-1,0]
vrange_theta=[-3, 0.5]
layout="vertical"
colormapToApply="jet"

#Rotate 90 degrees
# EXX = np.rot90(EXX,3)
# EYY = np.rot90(EYY,3)
# EXY = np.rot90(EXY,3)
# Rot = np.rot90(Rot,3)

ratio = np.sqrt(EXX.shape[1] / EXX.shape[0])
if layout == "square":
    figsize = (13 * ratio, 8 / ratio)
else:
    figsize = (20 * ratio, 10 / ratio)
```

```python
#Plot all maps
from mpl_toolkits.axes_grid1 import make_axes_locatable

fig = plt.figure(figsize=figsize)
ax1 = plt.subplot(2, 2, 1)
im1 = ax1.imshow(EXX, vmin=vrange_exx[0], vmax=vrange_exx[1], cmap=colormapToApply)
divider = make_axes_locatable(ax1)
cax1 = divider.append_axes("right", size="4%", pad=0.15)
cbar1 = plt.colorbar(im1, cax=cax1)
cbar1.ax.tick_params(labelsize=colorbarLabelSize) 
ax1.axes.xaxis.set_ticks([])
ax1.axes.yaxis.set_ticks([])
ax1.set_title(r"$\epsilon_{xx}$ (%)", size=titlesize)

ax2 = plt.subplot(2, 2,  2)
im2 = ax2.imshow(EYY, vmin=vrange_eyy[0], vmax=vrange_eyy[1], cmap=colormapToApply)
divider = make_axes_locatable(ax2)
cax2 = divider.append_axes("right", size="4%", pad=0.15)
cbar2 = plt.colorbar(im2, cax=cax2)
cbar2.ax.tick_params(labelsize=colorbarLabelSize) 
ax2.axes.xaxis.set_ticks([])
ax2.axes.yaxis.set_ticks([])
ax2.set_title(r"$\epsilon_{yy}$ (%)", size=titlesize)

ax3 = plt.subplot(2, 2,  3)
im3 = ax3.imshow(EXY, vmin=vrange_exy[0], vmax=vrange_exy[1], cmap=colormapToApply)
divider = make_axes_locatable(ax3)
cax3 = divider.append_axes("right", size="4%", pad=0.15)
cbar3 = plt.colorbar(im3, cax=cax3)
cbar3.ax.tick_params(labelsize=colorbarLabelSize) 
ax3.axes.xaxis.set_ticks([])
ax3.axes.yaxis.set_ticks([])
ax3.set_title(r"$\epsilon_{xy}$ (%)", size=titlesize)

ax4 = plt.subplot(2, 2, 4)
im4 = ax4.imshow(Rot, vmin=vrange_theta[0], vmax=vrange_theta[1], cmap=colormapToApply)
divider = make_axes_locatable(ax4)
cax4 = divider.append_axes("right", size="4%", pad=0.15)
cbar4 = plt.colorbar(im4, cax=cax4)
cbar4.ax.tick_params(labelsize=colorbarLabelSize) 
ax4.axes.xaxis.set_ticks([])
ax4.axes.yaxis.set_ticks([])
ax4.set_title(r"$\theta$ ($^\circ$)", size=titlesize)

fig.tight_layout(pad=1)
fig.savefig(StoringName+'_custom_all.png',bbox_inches='tight')
```
